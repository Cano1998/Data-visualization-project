```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
!pip install pandas seaborn
```

```python
# Load data
data = pd.read_csv('Cars93.csv')
```

### 1. Boxplot

```python
# Boxplot for revs per mile
list_cars = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']
cars = data[data['Manufacturer'].isin(list_cars)]
sns.boxplot(x='Manufacturer', y='Rev.per.mile', data=cars)
plt.show()
```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:** 
Suzuki


### 2. Histogram

```python
# Histogram of MPG
sns.histplot(data=data, x='MPG.city',kde=True, label='MPG.city')
sns.histplot(data=data, x='MPG.highway', kde=True, label='MPG.highway')
plt.legend(title='MPG city & MPG highway')
plt.xlabel('MPG')
plt.ylabel('Frecuency')
plt.title('MGP city VS. MGP highway')
plt.show()
```

**Histogram question:**  
Is it generally more fuel efficient to drive in the city or on the highway?

**Histogram answer:**
Highway

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
sns.lineplot(x='Wheelbase', y='Turn.circle', data=data)
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:** 
When the wheelbase is shorter, the turning radius is smaller. Therefore, when the wheelbase increases, the turning circle will be higher.


### 4. Barplot

```python
# Barplot for mean horsepower
sns.barplot(data, x='Type', y='Horsepower', hue='Type', legend=True)
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:**
Yes, it does. When the car is larager, it has more horsepower
